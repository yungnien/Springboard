## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
# testing to see if I understand how this works
# further populate tables created from nested element
json_normalize(data, record_path='counties',\
              meta=['state', 'shortname',['info','governor']], meta_prefix="meta_", record_prefix="record_")
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Inspecting the json object
import numpy as np
import regex

json_str = json.load(open('data/world_bank_projects.json'))
df = json_normalize(json_str, 
                   record_path=['mjtheme_namecode'],
                    meta=['id','countryshortname'],
                   record_prefix='theme_')

print(df.info())
```

```python
# let's clean up the data
# replace empty theme names with NaN
df = df.replace(r'^\s*$', np.nan, regex=True)
# sort by code and name so we can ffill NaN later
df = df.sort_values(['theme_code','theme_name'])
print("\nInfo after cleaning empty columns")
print(df.info())


# how many themes of each code do we have?
print('\nTheme Code Summary')
# this is a Series we will use later to verify ffill later
theme_code_summary = df.groupby('theme_code').size()
print(theme_code_summary)

# let's look at how many of each theme without theme name
print('\nEmpty Theme Name Summary')
print(df[df.theme_name.isnull()].groupby('theme_code').size())

```

```python
import regex
# forward fill empty theme names
df_filled = df.fillna(method='ffill')

# sanity check results against theme_code_summary - visual inspection
df_filled['theme_combo'] = df_filled.theme_code + df_filled.theme_name
print(df_filled.groupby('theme_name').size())
print(len(df_filled.groupby('theme_name').size()))
print("")
df_filled_combo = df_filled.groupby('theme_combo').size()
print(df_filled_combo)
print(len(df_filled_combo))


# programmatically test to see if number for each themes are still the same now that
# we have ffilled the theme names
# we do this by comparing with theme_code_summary numbers
print("\nCompare theme codes with ffilled theme names")
combo_indexes = df_filled_combo.index
pattern = r'\d+'
prog = regex.compile(pattern)
for idx in combo_indexes:
    m = prog.match(idx)
    summary_idx = m.group()
    print (f'comparing {df_filled_combo[idx]} with {theme_code_summary[summary_idx]}')
    assert df_filled_combo[idx] == theme_code_summary[summary_idx]

    


```

```python
# find 10 countries with most projects
# dedupe by project ID's
df_filled.drop_duplicates(subset=['id']).\
    groupby(['countryshortname']).size().sort_values(ascending=False).\
    iloc[:10]
```

```python
# find top 10 major project themes 'mjtheme_namecode'
df_filled.groupby(['theme_code','theme_name']).size().\
    sort_values(ascending=False)
```
